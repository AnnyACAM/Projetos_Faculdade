```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_5 = df[(df['armazens'] == 5) & (df['transport_capacity'] == 100) & (df['transport_interval'] == 1)]
df_100 = df[(df['armazens'] == 100) & (df['transport_capacity'] == 100) & (df['transport_interval'] == 1)]

plt.figure(figsize=(10,6))
plt.plot(df_5['pacotes'], df_5['tempo_exec_s'], marker='o', label='5 Armazéns')
plt.plot(df_100['pacotes'], df_100['tempo_exec_s'], marker='o', label='100 Armazéns')
plt.xlabel('Número de Pacotes')
plt.ylabel('Tempo de Execução (s)')
plt.title('Tempo de Execução vs Número de Pacotes\nComparação entre 5 e 100 Armazéns')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df = pd.read_csv("resultados_analise.csv")
df_armazens = df[(df['pacotes'] == 10000) & (df['transport_capacity'] == 100) & (df['transport_interval'] == 2)]

plt.figure(figsize=(10,6))
plt.plot(df_armazens['armazens'], df_armazens['tempo_exec_s'], marker='o')
plt.title('Tempo de Execução vs Número de Armazéns\n(Pacotes fixos = 10.000)')
plt.xlabel('Número de Armazéns')
plt.ylabel('Tempo de Execução (s)')
plt.grid(True)
plt.show()

# --- 2. Tempo de execução vs Número de pacotes (armazéns fixos = 400) ---
df_pacotes = df[(df['transport_capacity'] == 100) & (df['transport_interval'] == 1)]

plt.figure(figsize=(10,6))
plt.plot(df_pacotes['pacotes'], df_pacotes['tempo_exec_s'], marker='o', color='orange')
plt.title('Tempo de Execução vs Número de Pacotes\n(Armazéns fixos = 100)')
plt.xlabel('Número de Pacotes')
plt.ylabel('Tempo de Execução (s)')
plt.grid(True)
plt.show()

# --- 3. Efeito da capacidade e intervalo no tempo de execução ---
# Fixar armazéns=10, pacotes=1000 e variar capacidade e intervalo
df_contencao = df[(df['armazens'] == 10) & (df['pacotes'] == 1000)]

plt.figure(figsize=(10,6))
for capacity in sorted(df_contencao['transport_capacity'].unique()):
    subset = df_contencao[df_contencao['transport_capacity'] == capacity]
    plt.plot(subset['transport_interval'], subset['tempo_exec_s'], marker='o', label=f'Capacidade={capacity}')
plt.title('Tempo de Execução vs Intervalo de Transporte\n(Armazéns=10, Pacotes=1000)')
plt.xlabel('Intervalo de Transporte')
plt.ylabel('Tempo de Execução (s)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Define o estilo dos gráficos
sns.set(style="whitegrid")

# Cria o gráfico
plt.figure(figsize=(12, 7))

# Para cada valor único de pacotes, plota uma linha
for pacote_val in sorted(df['pacotes'].unique()):
    subset = df[df['pacotes'] == pacote_val]
    subset = subset.sort_values(by='armazens')
    plt.plot(subset['armazens'], subset['tempo_exec_s'], marker='o', label=f'{pacote_val} pacotes')

# Personaliza o gráfico
plt.title("Tempo de Execução vs Número de Armazéns\n(diferentes quantidades de pacotes)")
plt.xlabel("Número de Armazéns")
plt.ylabel("Tempo de Execução (s)")
plt.legend(title="Pacotes")
plt.tight_layout()

# Mostra o gráfico
plt.show()
```
